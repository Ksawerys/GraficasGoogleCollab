```python
import numpy as np
import matplotlib.pyplot as plt

# Definir las funciones de complejidad
def constante(n):
    return np.ones_like(n)

def logaritmica(n):
    return np.log(n)

def lineal(n):
    return n

def casi_lineal(n):
    return n * np.log(n)

def cuadratica(n):
    return n**2

def cubica(n):
    return n**3

def exponencial(n):
    return 2**n

# Crear un rango de valores para n
n_values = np.array([20, 50, 100, 200])

# Calcular los valores de las funciones de complejidad
constante_values = constante(n_values)
logaritmica_values = logaritmica(n_values)
lineal_values = lineal(n_values)
casi_lineal_values = casi_lineal(n_values)
cuadratica_values = cuadratica(n_values)
cubica_values = cubica(n_values)
exponencial_values = exponencial(n_values)

# Graficar las funciones en una escala logartmica
plt.figure(figsize=(10, 6))
plt.plot(n_values, constante_values, label='O(1) Constante')
plt.plot(n_values, logaritmica_values, label='O(log n) Logartmica')
plt.plot(n_values, lineal_values, label='O(n) Lineal')
plt.plot(n_values, casi_lineal_values, label='O(n log n) Casi Lineal')
plt.plot(n_values, cuadratica_values, label='O(n) Cuadrtica')
plt.plot(n_values, cubica_values, label='O(n) Cbica')
plt.plot(n_values, exponencial_values, label='O(2^n) Exponencial')

# Configurar la escala logartmica
plt.yscale('log')
plt.xscale('log')

# Aadir etiquetas y leyenda
plt.xlabel('Tamao de muestra (n)')
plt.ylabel('Tiempo de ejecucin')
plt.title('Comparativa de rdenes de Complejidad')
plt.legend()
plt.grid(True, which="both", ls="--")

# Mostrar la grfica
plt.show()
```
